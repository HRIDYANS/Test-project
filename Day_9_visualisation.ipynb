```python
filepath_auto = '/content/auto-mpg_edited.csv'
data_auto= pd.read_csv(filepath_auto)
data_auto.head()
```

```python
data_auto.columns
```

##Histogram

```python
plt.hist(data_auto['mpg'])
```

```python
plt.figure(figsize=(20,8))
plt.hist(data_auto['mpg'])
```

```python
plt.figure(figsize=(4,8))
plt.hist(data_auto['mpg'])
plt.title('Distribution of Milage (mpg)')
```

```python
plt.hist(data_auto['mpg'],color='red')
plt.title('Distribution of Milage (mpg)',
fontsize=20)
```

```python
plt.hist(data_auto['mpg'],color='red',rwidth=0.9)
plt.title('Distribution of Milage (mpg)',
fontsize=20)
```

```python
plt.hist(data_auto['mpg'],color='red',rwidth=0.9,cumulative=True)
plt.title('Distribution of Milage (mpg)',
fontsize=20)
```

Scatter plot

```python
plt.scatter(data_auto['horsepower'],data_auto['mpg'],s=10,c='r',marker ='*')
plt.title('plot of HP VS MPG')
plt.xlabel('HP')
plt.ylabel('MPG')
plt.show()
```

SUBPLOTS

```python
plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.scatter(data_auto['horsepower'],data_auto['mpg'],s=10,c='r',marker ='*')
plt.title('plot of HP VS MPG')
plt.xlabel('HP')
plt.ylabel('MPG')

plt.subplot(2,1,2)
plt.scatter(data_auto['horsepower'],data_auto['mpg'],s=10,c='r',marker ='*')
plt.title('plot of HP VS MPG')
plt.xlabel('HP')
plt.ylabel('MPG')
plt.show()
```
